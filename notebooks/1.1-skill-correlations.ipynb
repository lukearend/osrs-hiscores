```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from hiscores.data import load_hiscores_data, exclude_missing
```

```python
ranks, levels, xp = load_hiscores_data()
```

```python
attack_levels = xp['attack']
defence_levels = xp['defence']
```

```python
attack_levels = exclude_missing(attack_levels)
defence_levels = exclude_missing(defence_levels)
```

```python
skills = ('runecraft', 'farming')
first_skill = levels[skills[0]]
second_skill = levels[skills[1]]
concated = pd.concat([first_skill, second_skill], axis=1)
X = exclude_missing(np.array(concated))
```

```python
plt.figure(figsize=(8, 8))
noisy_x = X[:, 0] + np.random.rand(X.shape[0]) - 0.5
noisy_y = X[:, 1] + np.random.rand(X.shape[0]) - 0.5
keep_inds = np.random.permutation(X.shape[0])[:100000]
plt.plot(noisy_x[keep_inds], noisy_y[keep_inds], '.', markersize=1)
plt.yscale('linear')
plt.xscale('linear')
plt.ylim([0.5, 99.5])
plt.xlim([0.5, 99.5])
plt.show()
```
